```python
# Question 3
# Which year had the least correlation with ‘Element Code’?

# Fill in missing values to prevent error while calulating
df['Y2014'].fillna(df['Y2014'].mean(), inplace = True)
df['Y2015'].fillna(df['Y2014'].mean(), inplace = True)
df['Y2016'].fillna(df['Y2014'].mean(), inplace = True)
df['Y2017'].fillna(df['Y2014'].mean(), inplace = True)
df['Y2018'].fillna(df['Y2014'].mean(), inplace = True)

from scipy import stats

# Create an empty DataFrame to store the correlations and p-values
corr_df = pd.DataFrame(columns=['r', 'p-value'])

for col in df:  # Use this to loop through the insurance.csv DataFrame
  if pd.api.types.is_numeric_dtype(df[col]): # Only calculate r, p-value for the numeric columns
     r, p = stats.pearsonr(df['Element Code'], df[col])
     corr_df.loc[col] = [round(r, 3), round(p, 3)]

corr_df.sort_values(by=['r'], ascending=False)


corr_df
```
